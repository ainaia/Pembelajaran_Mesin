```python
df = pd.read_csv('/content/most-popular.csv')

# Membuat bar chart
plt.figure(figsize=(20,20))
colors = ['#FAEBD7' for _ in range(len(df))]
colors[-10] = '#D2691E'
plt.bar(df['show_title'], df['hours_viewed_first_28_days'], color=colors)
plt.title('Most Popular Netflix')
plt.xlabel('Judul')
plt.ylabel('Views')
plt.xticks(rotation=90, ha='right', fontsize=15)
plt.yticks(range(0,10,20))
plt.tight_layout()
plt.show()
```

Visualisasi diatas menunjukkan bahwa series netflix yang paling banyak ditonton adalah "Squid Game"
